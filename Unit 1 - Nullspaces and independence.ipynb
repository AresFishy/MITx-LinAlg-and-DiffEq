## Nullspace

The set of all solutions to a homogeneous linear system $\mathbf{A}\mathbf{x}= \mathbf{0}$ is called the <font color="blue">**nullspace of matrix A**</font>, and is denoted$\,  {\color{blue}{\mathrm{NS}(\mathbf{A})}}$

### Theorem: 

If **A** is a matrix, then the set of all solutions to the homogeneous linear system $Ax=0$ is a vector space, i.e. the nullspace is a subspace of $\mathbb{R}^n$

### <font color="orange">*Analogous theorem:*</font>

If $P(D)$ is a linear differential operator, then the set of solutions to the homogeneous ODE $P(D)x=0$ is a vector space of functions (i.e. a subspace of the set of all functions)

### Nullspace example:

Consider the matrix:

$$\mathbf {A} = \begin{pmatrix} 1&1&2 \\ 2&1&3 \\ 3&1&4 \\ 4&1&5 \end{pmatrix}$$

The system $Ax= \begin{pmatrix} 0 \\ 0 \\ 0 \\0 \end{pmatrix}$ has a number of solutions, such as:

$x = \begin{pmatrix} 0 \\ 0 \\ 0 \end{pmatrix}, and \ x = \begin{pmatrix} 1 \\ 1 \\ -1 \end{pmatrix}$

Since this is a homogeneous matrix, from knowing one solution we now know the entire nullspace! It is any multiple $c\begin{pmatrix} 1 \\ 1 \\ -1 \end{pmatrix}$. This is a line in $\mathbb{R}^3$! 

## Span of a set of vectors

The span of a set of vectors $v_1,...,v_n$ is the set of **all** linear combinations of $v_1,...,v_n$:

$${\color{blue}{\mathrm{Span}(\mathbf v_1,\ldots ,\mathbf v_ n)}}  = \{ \text {all vectors }c_1 \mathbf v_1 + \cdots + c_ n \mathbf v_ n, \text { where } c_1,\ldots ,c_ n \text { are scalars}\} .$$

The span is always a vector space. It provides a convenient notation for an infinite set of vector that form a subspace

E.g. if $v = \begin{pmatrix} 1 \\ 1 \end{pmatrix}$, then $\mathrm{Span}(\mathbf v)=\left\{ c \begin{pmatrix}  1 \\ 1\end{pmatrix}\right\}.$ This is the line $y=x$ in $\mathbb{R}^2$

Equivalently, the Span of a set of vectors which are multiples of each other is the span of the "first" vector, e.g.

$$\mathrm{Span}\left( \begin{pmatrix}  1 \\ 1\end{pmatrix}, \begin{pmatrix}  2 \\ 2\end{pmatrix} \right) = \mathrm{Span}\left( \begin{pmatrix}  1 \\ 1\end{pmatrix} \right).$$

### Note: 

Non-linear functions of vectors (e.g. ${x}^2$) do **not** constitute vector spaces!!

If we get the solutions of a homogeneous linear system in the form e.g.

$$c_1 \begin{pmatrix}  3/2 \\ 1 \\ 0 \\ 0 \\ 0 \\ 0 \end{pmatrix} + c_2 \begin{pmatrix}  -11/6 \\ 0 \\ -1/3 \\ 1 \\ 0 \\ 0 \end{pmatrix}+ c_3 \begin{pmatrix}  19/6 \\ 0 \\ 2/3 \\ 0 \\ 1 \\ 0 \end{pmatrix}.$$

Then, the nullspace $NS(\mathbf A)$ is 

$$\mathrm{NS}(\mathbf{A}) =\{ \text {the set of solutions to }\,  \mathbf{A}\mathbf{x}= \mathbf{0} \}  =\mathrm{Span}\left( \begin{pmatrix}  3/2 \\ 1 \\ 0 \\ 0 \\ 0 \\ 0 \end{pmatrix} , \begin{pmatrix}  -11/6 \\ 0 \\ -1/3 \\ 1 \\ 0 \\ 0 \end{pmatrix}, \begin{pmatrix}  19/6 \\ 0 \\ 2/3 \\ 0 \\ 1 \\ 0 \end{pmatrix} \right).$$


## Column space

The *column space* of matrix A is the span of its columns:

$$\mathrm{CS}(\mathbf{A}) = \mathrm{Span}\left( \text {columns of } \mathbf{A}\right) = \mathrm{Span}\left( \mathbf v_1,\dotsc \mathbf v_ n \right)$$

## Linear dependence and independence

In the example of $\mathrm{Span}\left( \begin{pmatrix}  1 \\ 1\end{pmatrix}, \begin{pmatrix}  2 \\ 2\end{pmatrix} \right) = \mathrm{Span}\left( \begin{pmatrix}  1 \\ 1\end{pmatrix} \right).$, it's easy to tell that the vector $\begin{pmatrix} 2 \\ 2 \end{pmatrix}$ is redundant, since it's in the span of $\begin{pmatrix} 1 \\ 1 \end{pmatrix}$. We say that these vectors are *linearly dependent*

But how do we generalize to 3 or more vectors?

For example. let's consider the span $\mathrm{Span}\left(\begin{pmatrix} 1 \\ 1 \\ 1\end{pmatrix}, \begin{pmatrix} 0 \\ 1 \\ 0\end{pmatrix}, \begin{pmatrix} 1 \\ 0 \\ 1\end{pmatrix} \right)$.

We can't write $\begin{pmatrix} 0 \\ 1 \\ 0 \end{pmatrix}$ and $\begin{pmatrix} 1 \\ 0 \\ 1 \end{pmatrix}$ as $c\begin{pmatrix} 1 \\ 1 \\ 1 \end{pmatrix}$. So, at most one vector is redundant.

However, vector $\begin{pmatrix} 1 \\ 0 \\ 1 \end{pmatrix}$ is in the span $\mathrm{Span}\left( \begin{pmatrix} 1 \\ 1 \\ 1\end{pmatrix} , \begin{pmatrix} 0 \\ 1 \\ 0\end{pmatrix} \right)$ since:

$$\begin{pmatrix}  1 \\ 0 \\ 1\end{pmatrix} = \begin{pmatrix} 1 \\ 1 \\ 1\end{pmatrix} - \begin{pmatrix} 0 \\ 1 \\ 0\end{pmatrix}.$$

### Geometric definition: 
A collection of vectors $\mathbf v_1, \mathbf v_2, \mathbf v_3, \dotsc , \mathbf v_ n$ is <font color="blue">linearly dependent</font> if there is at least one vector $\mathbf v_i$ that lies in the subspace spanned by the other vectors. 

### Algebraic definition:
A collection of vecttors $\mathbf v_1, \mathbf v_2, \mathbf v_3, \dotsc , \mathbf v_ n$ is <font color="blue">linearly dependent</font> if there exist scalars $\mathbf c_1, \dotsc, \mathbf c_n$ **not all zero** such that

$$c_1 \mathbf v_1 + \dotsb + c_ n \mathbf v_ n = 0.$$

I.e. if any combination of *not all zeros* gives the zero vector


Suppose we have 2 linearly independent vectors $\begin{pmatrix} 1 \\ 1 \\ 0 \end{pmatrix}$ and $\begin{pmatrix} 1 \\ -1 \\ 0 \end{pmatrix}$. These two vectors span a plane inside of $\mathbb{R}^3$. A third vector $\begin{pmatrix} a \\ b \\ c \end{pmatrix}$ is <font color="blue">linearly dependent</font> iff it lies in the plane spanned by the first two vectors, which in turn is true iff we can write it as a linear combination of the first two vectors.

So basically, in e.g. $\mathbb{R}^2$, any two independent non-zero vectors can "describe the entire plane", since any third vector can be expressed as a combination of the two vectors.

```julia
# Geometric definition concept check
using Plots
f(x,y,z) = x - 2y + z
# Vectors v1, v2 characterize the plane drawn by f(x,y,z) = 0 . 
# What value of c makes the vector (3; 1; c) dependent on v1, v2?
```

```julia
pyplot()
surface([(f(1,0,0), 0, 0), (0, f(0,1,0),0), (0, 0, f(0, 0, 1))])
```

The solution to $f(x,y,c)=0$ for $\begin{pmatrix} 3 \\ 1 \\ c \end{pmatrix}$ is $c=-1$. This makes the vector linearly dependent on the plane drawn by the equation.

### Independence and the nullspace 

If $\mathbf v_1 , \dotsc , \mathbf v_n$ are columns of *A*, they are <font color="orange">independent</font> if the nullspace of *A* is only the zero vector! 

They are <font color="blue">dependent</font> otherwise 

I.e. we can create a matrix whose columns are our vectors of interest, find its nullspace, and if it contains only the zero vector then the vectors are independent.

```julia
# Checking independence using the nullspace

A = [2 1 9; 3 2 11; -1 -3 8; 4 6 -4]
# We can cheat :D 
nullspace(A)
# Since it's zeros we can say that the column vectors of A are independent :)
```
